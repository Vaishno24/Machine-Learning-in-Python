# Box and Whisker Representation of Data

First, we arrange the given data in ascending order.
Now, we calculate the median of the data which is the main median.
Lower Quartile (Q1) is calculated as the median of the data from the left side of the main median.
Upper Quartile (Q3) is calculated as the median of the data from the right side of the main median.
MIN and MAX are the lowest and highest scores of the data excluding outliers.
Outliers is the point defined as a data point outside the whiskers of the box plot.
Outlier is usually outside 1.5 times the interquartile range above the upper quartile and below the lower quartile [Q1 - (1.5 * IQR) or Q3 + (1.5 * IQR)]
